```python
def factorial(n):
    if n == 0:
        return 1
    else:
        res = 1
        for i in range(1, n + 1):
            res *= i
        return res

# Example usage:
num = int(input("Enter a non-negative integer: "))
print(f"The factorial of {num} is {factorial(num)}")
```
