```python
a = 10
b = 5

# Addition
result_add = a + b
print(f"Addition: {a} + {b} = {result_add}")

# Subtraction
result_sub = a - b
print(f"Subtraction: {a} - {b} = {result_sub}")

# Multiplication
result_mul = a * b
print(f"Multiplication: {a} * {b} = {result_mul}")

# Division
result_div = a / b
print(f"Division: {a} / {b} = {result_div}")

# Floor Division
result_floor_div = a // b
print(f"Floor Division: {a} // {b} = {result_floor_div}")

# Modulus
result_mod = a % b
print(f"Modulus: {a} % {b} = {result_mod}")

# Exponentiation
result_exp = a ** b
print(f"Exponentiation: {a} ** {b} = {result_exp}")
```
